## 1. Profit Margin for Each product 

```python
Product_sales_df["Profit"] = Product_sales_df["Unit Price"] - Product_sales_df["Unit Cost"]
Product_sales_df["Margin (%)"] = (Product_sales_df["Profit"] / Product_sales_df["Unit Price"]) * 100
```

```python
## Classifying the profit margin into high and low margin

# Define a threshold â€” let's say 50% margin is the cutoff
Product_sales_df["Margin Category"] = Product_sales_df["Margin (%)"].apply(
    lambda x: "High Margin" if x >= 50 else "Low Margin"
)
```

```python

print("\nHigh-Margin Products:")
print(Product_sales_df[Product_sales_df["Margin Category"] == "High Margin"][["Product Name", "Unit Price", "Unit Cost", "Margin (%)"]])

print("\nLow-Margin Products:")
print(Product_sales_df[Product_sales_df["Margin Category"] == "Low Margin"][["Product Name", "Unit Price", "Unit Cost", "Margin (%)"]])
```

```python
# Filter high- and low-margin products
high_margin_df = Product_sales_df[Product_sales_df["Margin Category"] == "High Margin"].copy()
low_margin_df = Product_sales_df[Product_sales_df["Margin Category"] == "Low Margin"].copy()

# Add or ensure margin category columns are correct
high_margin_df["Margin Category"] = "High Margin"
low_margin_df["Margin Category"] = "Low Margin"
```

```python
## Visualize the high and low margin
# Add category label
high_margin_df = high_margin_df.copy()
low_margin_df = low_margin_df.copy()
high_margin_df["Margin Type"] = "High Margin"
low_margin_df["Margin Type"] = "Low Margin"

# Combine both DataFrames
combined_df = pd.concat([high_margin_df, low_margin_df])

# Sort for clarity
combined_df = combined_df.sort_values(by="Margin (%)", ascending=False)

# Plot
plt.figure(figsize=(16, 6))
sns.barplot(data=combined_df, x="Product Name", y="Margin (%)", hue="Margin Type")
plt.xticks(rotation=45, ha="right")
plt.title("Comparison of High and Low Margin Products")
plt.xlabel("Product Name")
plt.ylabel("Margin (%)")
plt.legend(title="Margin Category")
plt.tight_layout()
plt.show()
```

```python
## Target Gap Analysi to know if the sales target was met for each product 
```

```python
# Aggregate the sells 
Actual_sales = Product_sales_df.groupby("Division")["Sales"].sum().reset_index()
Actual_sales.rename(columns={"Sales": "Actual Sales"}, inplace=True)
```

```python
# Merge actual sales with targets
target_analysis = pd.merge(targets_df, Actual_sales, on="Division", how="left")
```

```python
target_analysis["Gap"] = target_analysis["Target"] - target_analysis["Actual Sales"]
target_analysis["Met Target"] = target_analysis["Gap"] <= 0
```

```python
print("Target Gap Analysis by Division:")
print(target_analysis)
```

```python
# Sort by actual sales for clarity
target_analysis = target_analysis.sort_values(by="Actual Sales", ascending=False)

# Plot with hue based on Met Target status
plt.figure(figsize=(10, 6))
sns.barplot(
    data=target_analysis,
    x="Division",
    y="Actual Sales",
    hue="Met Target",
    palette={True: "green", False: "red"}
)

plt.title("Division Sales Performance vs Target")
plt.xlabel("Division")
plt.ylabel("Actual Sales ($)")
plt.legend(title="Met Target", loc="upper right")
plt.tight_layout()
plt.show()
```
